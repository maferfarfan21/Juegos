```python
import pygame
import sys

pygame.init()

WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Juego en Pygame")

def mostrar_pantalla_inicial():
    font = pygame.font.Font(None, 36)
    tiempo_inicio = pygame.time.get_ticks()
    nombre_usuario = ""
    modo_juego = ""

    running = True
    while running:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        screen.fill((0, 0, 0))  # Fondo negro

        tiempo_actual = pygame.time.get_ticks()
        tiempo_transcurrido = (tiempo_actual - tiempo_inicio) // 1000  # segundos

        if tiempo_transcurrido < 10:
            texto = font.render(f"Pantalla de inicio - Tiempo restante: {30 - tiempo_transcurrido} segundos", True, (255, 255, 255))
            screen.blit(texto, (50, 50))
        else:
            if not nombre_usuario:
                texto_nombre = font.render("Ingrese su nombre:", True, (255, 255, 255))
                screen.blit(texto_nombre, (50, 50))

                for event in pygame.event.get():
                    if event.type == pygame.KEYDOWN:
                        if event.key == pygame.K_BACKSPACE:
                            nombre_usuario = nombre_usuario[:-1]
                        elif event.key == pygame.K_RETURN:
                            modo_juego = "Modo de Juego"  # Aquí debes definir tus modos de juego
                        else:
                            nombre_usuario += event.unicode

                texto_nombre_ingresado = font.render(nombre_usuario, True, (255, 255, 255))
                screen.blit(texto_nombre_ingresado, (50, 100))
            else:
                texto_modo = font.render(f"Seleccione un modo de juego: {modo_juego}", True, (255, 255, 255))
                screen.blit(texto_modo, (50, 50))
                # Aquí puedes agregar la lógica para seleccionar un modo de juego

        pygame.display.flip()

    pygame.quit()
    sys.exit()

if __name__ == "__main__":
    mostrar_pantalla_inicial()
```
